```python
# display image from directory using matplotlib
dir = "/blue/azarrinpar/ICG_AI/pytorch-CycleGAN-and-pix2pix/img_src/9-23/icg/"
images = sorted(os.listdir(dir))
image = cv2.imread(dir + images[0])
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(image)
# create histogram of brightness of image
plt.subplot(1,2,2)
plt.hist(image.ravel(), 256, [0, 256])
plt.show()
```

```python
# Create a histogram of the brightness of the first 1000 frames in the dir
brightness = []
for i in range(50):
    image = cv2.imread(dir + images[i])
    # brightness.append(np.mean(image))
    brightness += list(image.ravel())
plt.hist(brightness, 256, [0, 256])
plt.show()
```

```python
# plot an image
# set all pixels with brightness greater than 50 to red (255, 0, 0)
image = cv2.imread(dir + images[0])
plt.imshow(image)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread(dir + images[35000])

# Convert to grayscale to calculate brightness
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Create a mask where the average brightness is greater than 50
mask = gray_image > 120

# Set the pixels in the original image to red where the mask is true
# image[mask] = [0, 0, 255]
# Set the pixels in the original image to 25 where the mask is true
image[mask] = 25

# Display the modified image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
